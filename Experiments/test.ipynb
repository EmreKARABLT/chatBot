```python
#Loop through the train dataset and calculate how many images have one face
#and how many have more than one face
one_face = 0
more_than_one_face = 0
for i in range(len(dataset['train'])):
    print(i)
    if(i== 581 or i== 582 ):
        continue
    if len(dataset['train'][i]["faces"]["bbox"]) == 1:
        one_face += 1
    else:
        more_than_one_face += 1

print("Number of images with one face: ", one_face)
print("Number of images with more than one face: ", more_than_one_face)
```
